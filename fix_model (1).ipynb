```python
import os
os.chdir(r'C:\\Users\\karatmal narendar\\Desktop\\MLP\\research_paper_project')

import pickle
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Load vocab to get number of classes
with open('models/vocab.pkl', 'rb') as f:
    vocab = pickle.load(f)

num_classes = len(vocab)
MAX_TOKENS = 20000
print(f'Number of classes: {num_classes}')

# Rebuild exact same model architecture
def build_model(num_classes):
    inputs = keras.Input(shape=(MAX_TOKENS,), name='text_input')
    x = layers.Dense(512, activation='relu')(inputs)
    x = layers.Dropout(0.3)(x)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.Dropout(0.3)(x)
    x = layers.Dense(128, activation='relu')(x)
    outputs = layers.Dense(num_classes, activation='sigmoid', name='outputs')(x)
    model = keras.Model(inputs, outputs)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

new_model = build_model(num_classes)
print('New model built!')
```

```python
# Load weights from old model
new_model.load_weights('models/model.h5')
print('Weights loaded!')

# Save in new format
new_model.save('models/model_new.keras')
print('Done! model_new.keras saved!')
```
